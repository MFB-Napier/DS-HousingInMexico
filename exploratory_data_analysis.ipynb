```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

# **Import Data**

```python
df=pd.read_csv("/content/mexico-city-real-estate.csv")
```

```python
print("df type:", type(df))
print("df shape:", df.shape)
df.head()
```

# **Location Data: "lat" and "lon"**

--- See where in Mexico the houses are located by using the "lat" and "lon" columns. Since Latitude and Longitude are based on the coordinate system, to visualize them a scatter plot on top of a map is created, using the scatter_mapbox from plotly library.



```python
#Use plot express to create figure
fig = px.scatter_mapbox(
    df,
    center={"lat":19.310330, "lon": -99.068557}, #Map will be centered in mexico city with these coordinates
    width=600,
    height=600,
    hover_data=["price_aprox_usd"] #Display price when hovering mouse over house
)

#Add mapbox_style to figure layout
fig.update_layout(mapbox_style="open-street-map")

#show figure
fig.show()
```
